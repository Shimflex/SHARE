```python
# Calculate simple percentage returns from a list of prices
prices = [100.0, 102.5, 101.0, 103.0, 105.5]
returns = [] # Start with an empty list to store results

# Loop through prices starting from the second price
for i in range(1, len(prices)):
  # Calculate percentage return: (current_price - previous_price) / previous_price
  daily_return = (prices[i] - prices[i-1]) / prices[i-1] * 100
  returns.append(daily_return) # Add the calculated return to our list

print(f"Calculated Returns (%): {returns}")
# Output: Calculated Returns (%): [2.5, -1.4634146341463414, 1.9801980198019802, 2.4271844660194173]
```

```python
# Define a simple function to calculate potential claim payout
def calculate_payout(policy_limit, deductible, loss_amount):
  """Calculates the insurance payout amount."""
  if loss_amount <= deductible:
    return 0.0 # Payout is zero if loss is below deductible
  else:
    potential_payout = loss_amount - deductible
    # Payout cannot exceed the policy limit
    return min(potential_payout, policy_limit)

# Use the function
limit = 5000
deduct = 500
loss = 3200

payout = calculate_payout(limit, deduct, loss)
print(f"Calculated Payout: ${payout:.2f}") # Output: Calculated Payout: $2700.00

loss_small = 400
payout_small = calculate_payout(limit, deduct, loss_small)
print(f"Calculated Payout (Small Loss): ${payout_small:.2f}") # Output: Calculated Payout (Small Loss): $0.00
```

```python
import pandas as pd
import numpy as np

# Load the raw claims data
# Make sure 'claims_data_raw.csv' is in a 'datasets' folder
file_path = 'datasets/claims_data_raw.csv'
df = pd.read_csv(file_path)

# Display initial info and first few rows
print("Initial Data Info:")
df.info()
print("\nInitial Data Head:")
print(df.head())
```

```python
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# Fill missing ClaimAmount with 0 (example)
# df['ClaimAmount'] = df['ClaimAmount'].fillna(0)

# Fill missing Region with 'Unknown'
df['Region'] = df['Region'].fillna('Unknown')

# Fill missing ClaimDate with a specific placeholder or forward/backward fill
# For this example, let's fill with a placeholder date, though ffill/bfill are common
df['ClaimDate'] = df['ClaimDate'].fillna('1900-01-01') # Placeholder

print("\nMissing values after initial fill:")
print(df.isnull().sum())
```

```python
# Clean and convert ClaimAmount
# Remove '$' and ',' before converting
df['ClaimAmount'] = df['ClaimAmount'].str.replace('[$,]', '', regex=True).str.strip()
# Use errors='coerce' to turn problematic conversions into NaN
df['ClaimAmount'] = pd.to_numeric(df['ClaimAmount'], errors='coerce')
# Now fill NaNs created during coercion (if any), or previously existing ones
mean_claim_amount = df['ClaimAmount'].mean() # Calculate mean *after* conversion
df['ClaimAmount'] = df['ClaimAmount'].fillna(mean_claim_amount)
print("\nClaimAmount column type after cleaning:", df['ClaimAmount'].dtype)

# Clean and convert SettlementAmount (handling mixed types)
df['SettlementAmount'] = pd.to_numeric(df['SettlementAmount'], errors='coerce') # Turns non-numbers into NaN
median_settlement = df['SettlementAmount'].median() # Calculate median
df['SettlementAmount'] = df['SettlementAmount'].fillna(median_settlement)
print("SettlementAmount column type after cleaning:", df['SettlementAmount'].dtype)

# Convert ClaimDate to datetime objects
# errors='coerce' will turn unparseable dates into NaT (Not a Time)
df['ClaimDate'] = pd.to_datetime(df['ClaimDate'], errors='coerce')
# Fill any NaT values if needed (e.g., with the earliest or latest date, or drop)
df = df.dropna(subset=['ClaimDate']) # Example: drop rows with unparseable dates
print("ClaimDate column type after cleaning:", df['ClaimDate'].dtype)

print("\nInfo after type conversions:")
df.info()
```

```python
# Clean PolicyID whitespace
df['PolicyID'] = df['PolicyID'].str.strip()

# Clean PolicyType: strip spaces and convert to lower case
df['PolicyType'] = df['PolicyType'].str.strip().str.lower()
print("\nUnique Policy Types after cleaning:", df['PolicyType'].unique())

# Clean Region: strip spaces and convert to title case (optional, depends on desired format)
df['Region'] = df['Region'].str.strip().str.title()
print("Unique Regions after cleaning:", df['Region'].unique())

print("\nData head after string cleaning:")
print(df.head())
```

```python
# Calculate claim-settlement difference
df['SettlementDifference'] = df['ClaimAmount'] - df['SettlementAmount']

# Extract claim year
df['ClaimYear'] = df['ClaimDate'].dt.year

print("\nData head after transformations:")
print(df.head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
import os
```

```python
# Load the dataset
portfolio_df = pd.read_csv("datasets/portfolio_returns.csv", parse_dates=["Date"], index_col="Date")
```

```python
# Load the dataset
file_path = "datasets/portfolio_returns.csv"
try:
    portfolio_df = pd.read_csv(file_path, parse_dates=["Date"], index_col="Date")
    print("Real stock returns data loaded successfully.")
    print(portfolio_df.head())
except FileNotFoundError:
    print(f"Error: '{file_path}' not found.")
    print("Please ensure the file exists in the 'datasets' folder.")
    portfolio_df = None
except Exception as e:
    print(f"An error occurred: {e}")
    portfolio_df = None
```

```python
# Calculate VaR for a selected stock
confidence_level = 0.95
var_level = 1 - confidence_level

for stock_name in portfolio_df.columns:
    stock_returns = portfolio_df[stock_name].dropna()
    VaR_95 = stock_returns.quantile(var_level)
    print(f"{stock_name} - 1-Day Value at Risk (VaR) at 95% confidence: {VaR_95:.4f}")
    print(f"Interpretation: Based on historical data, we are 95% confident that the worst daily loss")
    print(f"for {stock_name} will not exceed {abs(VaR_95)*100:.2f}%.")
    print("-")
```

```python
import matplotlib.pyplot as plt

# Set up a grid of subplots
num_stocks = len(portfolio_df.columns)
rows = 3  # Organize plots in 3 columns
fig, axes = plt.subplots(rows, 3, figsize=(15, rows * 5))
axes = axes.flatten()  # Flatten the axes array to make indexing easier

confidence_level = 0.95
var_level = 1 - confidence_level

for i, stock_name in enumerate(portfolio_df.columns):
    stock_returns = portfolio_df[stock_name].dropna()
    VaR_95 = stock_returns.quantile(var_level)

    # Plotting the distribution of returns
    ax = axes[i]  # Get the axis for the current plot
    ax.hist(stock_returns, bins=50, alpha=0.7, color='blue', edgecolor='black')
    
    # Add a vertical line for VaR
    ax.axvline(x=VaR_95, color='red', linestyle='--', label=f'VaR 95%: {VaR_95:.4f}')
    
    # Add labels and title
    ax.set_title(f"Returns Distribution for {stock_name}")
    ax.set_xlabel('Daily Return')
    ax.set_ylabel('Frequency')
    ax.legend()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Assume equal weights for simplicity
num_assets = len(portfolio_df.columns)
weights = np.array([1/num_assets] * num_assets) # Equal weights: [0.25, 0.25, 0.25, 0.25]

# Calculate mean daily returns for each asset
mean_daily_returns = portfolio_df.mean()

# Calculate the covariance matrix of daily returns
cov_matrix = portfolio_df.cov()

# Calculate expected portfolio daily return
portfolio_return_daily = np.sum(mean_daily_returns * weights)

# Calculate expected portfolio daily volatility (standard deviation)
portfolio_volatility_daily = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))

# Often helpful to annualize (assuming ~252 trading days per year)
annual_factor = 252
portfolio_return_annual = portfolio_return_daily * annual_factor
portfolio_volatility_annual = portfolio_volatility_daily * np.sqrt(annual_factor)

print("\n--- Equally Weighted Portfolio ---")
print(f"Expected Annual Return: {portfolio_return_annual:.4f} ({portfolio_return_annual*100:.2f}%)")
print(f"Expected Annual Volatility (Risk): {portfolio_volatility_annual:.4f} ({portfolio_volatility_annual*100:.2f}%)")
```

```python
num_portfolios = 5000 # Number of random portfolios to simulate
results = np.zeros((3, num_portfolios)) # Store results: [return, volatility, sharpe ratio]
all_weights = np.zeros((num_portfolios, num_assets)) # Store weights for each portfolio

print(f"\nSimulating {num_portfolios} random portfolios...")

for i in range(num_portfolios):
    # Generate random weights that sum to 1
    weights = np.random.random(num_assets)
    weights /= np.sum(weights) # Normalize weights to sum to 1
    all_weights[i,:] = weights

    # Calculate portfolio return and volatility (annualized)
    port_return = np.sum(mean_daily_returns * weights) * annual_factor
    port_volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix * annual_factor, weights)))

    # Store results (we'll ignore Sharpe Ratio for now to keep it simple)
    results[0,i] = port_return
    results[1,i] = port_volatility
    results[2,i] = port_return / port_volatility # Basic Sharpe Ratio (assuming risk-free rate = 0)

# Convert results array to DataFrame for easier plotting
results_frame = pd.DataFrame(results.T, columns=['Return', 'Volatility', 'Sharpe']) # Add 'Sharpe' if calculated

# Plot the Efficient Frontier (simulation)
plt.figure(figsize=(8, 5))
plt.scatter(results_frame.Volatility, results_frame.Return, c=results_frame.Return/results_frame.Volatility, marker='o', s=10, alpha=0.6, cmap='viridis') # Color by Sharpe Ratio approximation
plt.title('Efficient Frontier Simulation')
plt.xlabel('Annualized Volatility (Risk)')
plt.ylabel('Annualized Return')
plt.colorbar(label='Return/Volatility Ratio (Approximation)')
plt.grid(True)
plt.show()
```

```python
# --- Simulation Parameters ---

# Portfolio characteristics
n_loans = 1000         # Number of loans in the portfolio
ead_per_loan = 50000   # Exposure At Default for each loan ($)
lgd = 0.40           # Loss Given Default (as a fraction, 40%)
pd_avg = 0.02        # Average Probability of Default for each loan (2%)

# Simulation settings
n_simulations = 10000  # Number of simulation runs (e.g., simulating 10,000 possible yearly outcomes)

# Calculate loss amount if a single loan defaults
loss_per_default = lgd * ead_per_loan

print(f"--- Portfolio & Simulation Setup ---")
print(f"Number of loans: {n_loans}")
print(f"Exposure per loan: ${ead_per_loan:,.2f}")
print(f"Loss Given Default (LGD): {lgd:.0%}")
print(f"Average Probability of Default (PD): {pd_avg:.1%}")
print(f"Loss per default event: ${loss_per_default:,.2f}")
print(f"Number of simulations (years): {n_simulations}")
```

```python
# --- Run the Simulation ---

# Simulate the number of defaults in each simulation year
# using the Binomial distribution:
# n = number of trials (loans)
# p = probability of success (default) in each trial
# size = number of simulation runs
print(f"\nRunning {n_simulations} simulations...")

num_defaults_per_sim = np.random.binomial(n=n_loans, p=pd_avg, size=n_simulations)

# Calculate the total portfolio loss for each simulation run
total_losses = num_defaults_per_sim * loss_per_default

print("Simulation complete.")
```

```python
# --- Analyze Results ---

# 1. Plot the distribution of simulated total losses
plt.figure(figsize=(10, 6))
plt.hist(total_losses, bins=50, edgecolor='black', alpha=0.7)
plt.title(f'Distribution of Simulated Annual Portfolio Credit Losses ({n_simulations:,} runs)')
plt.xlabel('Total Annual Loss ($)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Optional: Format x-axis labels for better readability (e.g., add commas)
formatter = FuncFormatter(lambda x, p: format(int(x), ','))
plt.gca().xaxis.set_major_formatter(formatter)
plt.xticks(rotation=30) # Rotate labels slightly if they overlap
plt.tight_layout() # Adjust layout
plt.show()

# 2. Calculate and display key statistics
# Analytical Expected Loss (for comparison)
expected_loss_analytic = n_loans * pd_avg * loss_per_default

# Simulation results
mean_simulated_loss = np.mean(total_losses)
median_simulated_loss = np.median(total_losses)
std_dev_simulated_loss = np.std(total_losses)
min_simulated_loss = np.min(total_losses)
max_simulated_loss = np.max(total_losses)

# Value at Risk (VaR) - common risk measures from percentiles
var_95 = np.percentile(total_losses, 95)  # 95% VaR
var_99 = np.percentile(total_losses, 99)  # 99% VaR
var_99_9 = np.percentile(total_losses, 99.9) # 99.9% VaR

# Unexpected Loss (UL) = VaR - Expected Loss
# Often calculated using a specific VaR level, e.g., 99% or 99.9%
unexpected_loss_99 = var_99 - mean_simulated_loss
unexpected_loss_99_9 = var_99_9 - mean_simulated_loss

print(f"\n--- Simulation Results ---")
print(f"Analytical Expected Loss:          ${expected_loss_analytic:,.2f}")
print(f"Simulated Mean Loss (EL):        ${mean_simulated_loss:,.2f}")
print(f"Simulated Median Loss:             ${median_simulated_loss:,.2f}")
print(f"Simulated Std Dev of Loss:       ${std_dev_simulated_loss:,.2f}")
print(f"Minimum Simulated Loss:          ${min_simulated_loss:,.2f}")
print(f"Maximum Simulated Loss:          ${max_simulated_loss:,.2f}")
print(f"\nValue at Risk (VaR):")
print(f"  95.0% VaR (Loss won't exceed this 95% of the time): ${var_95:,.2f}")
print(f"  99.0% VaR (Loss won't exceed this 99% of the time): ${var_99:,.2f}")
print(f"  99.9% VaR (Loss won't exceed this 99.9% of the time): ${var_99_9:,.2f}")
print(f"\nUnexpected Loss (UL):")
print(f"  UL at 99.0% (99% VaR - Mean):    ${unexpected_loss_99:,.2f}")
print(f"  UL at 99.9% (99.9% VaR - Mean):  ${unexpected_loss_99_9:,.2f}")
```

```python
# Create a Kaplan-Meier fitter instance
kmf = KaplanMeierFitter()

# Fit the model using duration and event columns
# 'duration' = Time observed
# 'event' = 1 if event occurred (e.g., death), 0 if censored (e.g., still active/lapsed)
kmf.fit(durations=survival_data['duration'], event_observed=survival_data['event'])

# Plot the survival function
plt.figure(figsize=(8, 5))
kmf.plot_survival_function()
plt.title('Kaplan-Meier Estimate of Policyholder Survival')
plt.xlabel('Duration (Years)')
plt.ylabel('Survival Probability')
plt.grid(True)
plt.show()

# You can also get the median survival time
median_survival = kmf.median_survival_time_
print(f"\nMedian survival time (years): {median_survival:.2f}")
```

```python
# Load the life table dataset
life_table = pd.read_csv('datasets/life_table_reserves.csv')

print("\nLife Table Data:")
print(life_table.head())
```

```python
import numpy as np

# Age for which to calculate remaining expectancy
start_age = 65

# Filter the table from the start age onwards
future_life_table = life_table[life_table['age'] >= start_age].copy()

# Number surviving at the start age
lx_start = future_life_table.iloc[0]['lx']

if lx_start > 0:
    # Calculate Tx = Sum of lx for all ages >= start_age
    # This represents the total person-years lived beyond the start age
    Tx = future_life_table['lx'].sum()

    # Crude estimate of remaining life expectancy: ex = Tx / lx_start
    # (Note: More accurate methods exist, often averaging lx and lx+1)
    remaining_expectancy = Tx / lx_start
    print(f"\nApproximate remaining life expectancy at age {start_age}: {remaining_expectancy:.2f} years")
else:
    print(f"\nCannot calculate life expectancy at age {start_age} (lx=0).")

# --- Visualization ---
plt.figure(figsize=(8, 5))

# Plot the survival curve (lx vs age)
plt.plot(life_table['age'], life_table['lx'], label='Number Surviving (lx)', color='blue')

# Shade the area representing Tx (total person-years lived beyond start_age)
if not future_life_table.empty:
    plt.fill_between(future_life_table['age'], 0, future_life_table['lx'], color='skyblue', alpha=0.5, label=f'Total Person-Years > Age {start_age} (Tx proxy)')

# Mark the start age and lx at start age
if lx_start > 0:
    plt.scatter([start_age], [lx_start], color='red', zorder=5)
    plt.text(start_age, lx_start, f'  Age {start_age}\n  lx = {int(lx_start)}', verticalalignment='bottom', color='red')
    # Display the calculated expectancy
    plt.text(0.6, 0.7, f'Remaining Expectancy\nat Age {start_age}:\n{remaining_expectancy:.2f} years',
                transform=plt.gca().transAxes, fontsize=12, bbox=dict(boxstyle='round,pad=0.5', fc='wheat', alpha=0.7))

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Number Surviving (lx)')
plt.title('Survival Curve and Life Expectancy Calculation')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.ylim(bottom=0) # Ensure y-axis starts at 0
plt.xlim(left=0)  # Ensure x-axis starts at 0

# Show the plot
plt.show()

```

```python
# Define parameters
start_age_reserve = 70
interest_rate = 0.05 # Example annual discount rate

# Filter table from the reserve start age
reserve_table = life_table[life_table['age'] >= start_age_reserve].copy()

# Calculate dx = number dying between age x and x+1 (lx - lx+1)
reserve_table['lx_plus_1'] = reserve_table['lx'].shift(-1).fillna(0) # Get lx for next age
reserve_table['dx'] = reserve_table['lx'] - reserve_table['lx_plus_1']
reserve_table['dx'] = reserve_table['dx'].clip(lower=0) # Ensure non-negative deaths

# Calculate probability of death at age x (used for checking, qx is already there)
# reserve_table['qx_calc'] = reserve_table['dx'] / reserve_table['lx']

# Time periods (years from start_age_reserve)
reserve_table['t'] = reserve_table['age'] - start_age_reserve + 1 # Discount period (end of year)

# Present Value factor v^t = 1 / (1 + i)^t
reserve_table['pv_factor'] = 1 / ((1 + interest_rate) ** reserve_table['t'])

# Expected Claims at age x = dx * avg_claim_amount
reserve_table['expected_claims'] = reserve_table['dx'] * reserve_table['avg_claim_amount']

# Present Value of Expected Claims at age x
reserve_table['pv_expected_claims'] = reserve_table['expected_claims'] * reserve_table['pv_factor']

# Total Present Value of Future Benefits for the cohort starting at start_age_reserve
total_pvfb = reserve_table['pv_expected_claims'].sum()

# Reserve per policyholder at start_age_reserve
lx_start_reserve = reserve_table.iloc[0]['lx']
reserve_per_policyholder = total_pvfb / lx_start_reserve

print(f"\nSimplified Reserve Calculation (starting age {start_age_reserve}):")
# print(reserve_table[['age', 'lx', 'dx', 'avg_claim_amount', 't', 'pv_factor', 'pv_expected_claims']].head())
print(f"Total Present Value of Future Benefits for cohort: ${total_pvfb:,.2f}")
print(f"Reserve per policyholder at age {start_age_reserve}: ${reserve_per_policyholder:,.2f}")
```

```python
import pandas as pd

# Load the claims data
file_path = 'datasets/claims_dashboard_data.csv'
claims_df = pd.read_csv(file_path)

# Display the first few rows to understand the structure
print("Claims Data:")
print(claims_df.head())
```

```python
import dash
from dash import dcc, html # Dash Core Components & HTML Components
import plotly.express as px
import pandas as pd
from dash.dependencies import Input, Output

# --- 1. Load Data ---
# (Repeat loading inside the script for self-containment)
file_path = 'datasets/claims_dashboard_data.csv'
claims_df = pd.read_csv(file_path)

# --- 2. Initialize the Dash App ---
app = dash.Dash(__name__)
app.title = "Claims Analysis Dashboard" # Browser tab title

# --- 3. Define the App Layout ---
app.layout = html.Div(children=[
    # Page Title
    html.H1(children='Insurance Claims Analysis'),

    # Subtitle/Description
    html.Div(children='''
        Analyze total claim amounts by policy type and region.
    '''),

    # Region Filter Dropdown
    html.Label('Select Region:'),
    dcc.Dropdown(
        id='region-dropdown',
        options=[{'label': 'All Regions', 'value': 'All'}] + # Add 'All' option
                [{'label': region, 'value': region} for region in claims_df['region'].unique()],
        value='All', # Default value
        clearable=False, # Prevent the user from clearing the selection
        style={'width': '50%'} # Adjust dropdown width
    ),

    # Bar Chart (placeholder, will be updated by callback)
    dcc.Graph(
        id='claims-by-policy-type-graph'
    )
])

# --- 4. Define Callbacks for Interactivity ---
# This function links the dropdown (Input) to the graph (Output)
@app.callback(
    Output('claims-by-policy-type-graph', 'figure'), # The graph's figure property
    Input('region-dropdown', 'value')               # The dropdown's selected value
)
def update_graph(selected_region):
    # Filter data based on dropdown selection
    if selected_region == 'All':
        filtered_df = claims_df # Use all data if 'All' is selected
    else:
        filtered_df = claims_df[claims_df['region'] == selected_region]

    # Group by policy type and sum claim amounts
    grouped_data = filtered_df.groupby('policy_type')['claim_amount'].sum().reset_index()

    # Create the bar chart using Plotly Express
    fig = px.bar(grouped_data,
                 x='policy_type',
                 y='claim_amount',
                 title=f'Total Claim Amount by Policy Type ({selected_region})',
                 labels={'policy_type': 'Policy Type', 'claim_amount': 'Total Claim Amount ($)'},
                 template='plotly_white' # Use a clean template
                )

    # Update layout for better readability
    fig.update_layout(
        title_x=0.5, # Center title
        xaxis_title="Policy Type",
        yaxis_title="Total Claim Amount ($)"
    )

    return fig # Return the updated figure to the graph component

# --- 5. Run the App Server ---
if __name__ == '__main__':
    app.run(debug=True) # debug=True allows auto-reloading on code changes
```
