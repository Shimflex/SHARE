## Loops: Repeating Actions
### For Loops: Iterating over Collections


```python
portfolio = [
    {"ticker": "AAPL", "quantity": 10, "price": 150},
    {"ticker": "MSFT", "quantity": 5, "price": 250},
    {"ticker": "GOOG", "quantity": 2, "price": 2000},
]

for stock in portfolio:
    print(f"Ticker: {stock['ticker']}, Value: {stock['quantity'] * stock['price']}")

# Using range() for numerical loops
for i in range(5):  # Iterates from 0 to 4
    print(i)

# Iterating with enumerate to get both index and value
for index, stock in enumerate(portfolio):
    print(f"Stock at index {index}: {stock['ticker']}")

# Using break to exit a loop early
for stock in portfolio:
    if stock['ticker'] == 'MSFT':
        print("Found Microsoft!")
        break # exit from for loop
    print(f"Checking {stock['ticker']}")

# Using continue to skip to the next iteration
for stock in portfolio:
    if stock['ticker'] == 'MSFT':
        continue # skip the rest of current loop
    print(f"Processing {stock['ticker']}") 
```

### While Loops: Repeating Until a Condition is Met


```python
count = 0
while count < 10:
    print(count)
    count += 1  # Important: Update the condition to avoid an infinite loop!

# While loop with break
balance = 1000
target_balance = 2000
interest_rate = 0.05
years = 0

while balance < target_balance:
    interest = balance * interest_rate
    balance += interest
    years += 1
    if years > 20: # prevent infinite loop
        print("Target not reached within 20 years.")
        break

print(f"Balance reached ${balance:.2f} in {years} years.")
```

## Replicating Excel's SUMIFS Functionality


```python
transactions = [
    {"ticker": "AAPL", "sector": "Tech", "value": 1500},
    {"ticker": "MSFT", "sector": "Tech", "value": 2500},
    {"ticker": "JPM", "sector": "Finance", "value": 1000},
    {"ticker": "BAC", "sector": "Finance", "value": 500},
]

total_tech_value = 0
for transaction in transactions:
    if transaction["sector"] == "Tech":
        total_tech_value += transaction["value"]

print(f"Total value of Tech sector transactions: ${total_tech_value}")


# More complex SUMIFS-like logic
total_value_tech_over_1000 = 0
for transaction in transactions:
    if transaction["sector"] == "Tech" and transaction["value"] > 1000:
        total_value_tech_over_1000 += transaction["value"]

print(f"Total value of Tech sector transactions over $1000: ${total_value_tech_over_1000}")
```

## Function Arguments and Return Values


```python
def add(x, y):
    """Adds two numbers and returns the sum."""
    return x + y

result = add(5, 3)  # result will be 8
print(result)

def calculate_loan_payment(principal, interest_rate, term):
    """Calculates the monthly loan payment."""
    monthly_interest_rate = interest_rate / 12
    num_payments = term * 12
    payment = (principal * monthly_interest_rate * (1 + monthly_interest_rate)**num_payments) / ((1 + monthly_interest_rate)**num_payments - 1)
    return payment

loan_payment = calculate_loan_payment(10000, 0.05, 5)
print(f"Monthly loan payment: ${loan_payment:.2f}")

# A function can return multiple values as a tuple
def get_stock_data(ticker):
    # In a real application, you would fetch data from a source like a database or API
    # Here, we'll just return some sample data
    if ticker == "AAPL":
        return 150, 10, "Tech"
    elif ticker == "MSFT":
        return 250, 5, "Tech"
    else:
        return None, None, None

price, quantity, sector = get_stock_data("AAPL")
print(f"Price: {price}, Quantity: {quantity}, Sector: {sector}")
```

## Reusable Financial Functions


```python
def net_present_value(cash_flows, discount_rate):
    """Calculates the net present value of a series of cash flows."""
    npv = 0
    for i, cash_flow in enumerate(cash_flows):
        npv += cash_flow / (1 + discount_rate)**i
    return npv

cash_flows = [-100, 20, 30, 40, 50]
discount_rate = 0.1
npv = net_present_value(cash_flows, discount_rate)
print(f"Net Present Value: ${npv:.2f}")

def internal_rate_of_return(cash_flows):
    """Calculates the internal rate of return using a simple iterative approach."""
    # This is a simplified example and might not converge for all cases.
    # For more robust IRR calculation, consider using libraries like NumPy or SciPy.
    rate = 0.1  # Initial guess
    for _ in range(100): # try up to 100 times to find the IRR
        npv = net_present_value(cash_flows, rate)
        if abs(npv) < 0.001:  # Convergence check
            return rate
        if npv > 0:
          rate += 0.01
        else:
          rate -= 0.01

    return "IRR calculation did not converge."  # Indicate failure to converge

irr = internal_rate_of_return(cash_flows)
print(f"Internal Rate of Return: {irr}")

def calculate_future_value(present_value, interest_rate, periods):
  """Calculates the future value of an investment."""
  future_value = present_value * (1 + interest_rate)**periods
  return future_value

future_value = calculate_future_value(1000, 0.05, 10)
print(f"Future Value: ${future_value:.2f}") 
```

```python
import pandas as pd

data = {'Date': ['2024-01-01', '2024-01-02', '2024-01-01', '2024-01-02'],
        'Ticker': ['AAPL', 'AAPL', 'MSFT', 'MSFT'],
        'Price': [150.25, 151.18, 250.75, 257.11]}

df = pd.DataFrame(data)

# Selecting a column (Series object)
tickers = df['Ticker']
print(tickers)
print(type(tickers))  # a pandas Series

# Selecting multiple columns (DataFrame object)
subset = df[['Ticker', 'Price']]
print(subset)
print(type(subset))  # a pandas DataFrame

# Selecting a row (using .loc for label-based indexing - returns a Series)
first_row = df.loc[0]
print(first_row)

# Selecting a cell (using .loc)
price_aapl = df.loc[0, 'Price']
print(price_aapl)

# Selecting rows and columns (using .loc - returns a DataFrame)
subset2 = df.loc[[0, 2], ['Ticker', 'Price']]
print(subset2)

# Selecting rows and columns (using .iloc for integer-based indexing - returns a DataFrame)
subset3 = df.iloc[[0, 2], [1, 2]]  # Select rows 0 and 2, and columns 1 (Ticker) and 2 (Price)
print(subset3)

# Boolean indexing (filtering)
high_price_stocks = df[df['Price'] > 200]
print(high_price_stocks)

# More complex boolean indexing (using logical operators)
tech_stocks_over_100 = df[(df['Ticker'].isin(['AAPL', 'MSFT'])) & (df['Price'] > 100)]
print(tech_stocks_over_100)

# Setting values (modifying the DataFrame in place)
df.loc[0, 'Price'] = 155  # Change the price of AAPL
print(df)

# Using .at and .iat for very fast scalar access (single value)
df.at[0, 'Price'] = 152  # Label-based
df.iat[0, 2] = 151  # Integer-based
print(df) 
```

# Chapter 8. Data Cleaning: Taming the Mess
## Handling Missing Values


```python
import pandas as pd
import numpy as np

data = {'Ticker': ['AAPL', 'MSFT', 'GOOG', 'AMZN', 'TSLA'],
        'Price': [150.25, 250.75, np.nan, 120.50, 700],  # NaN value
        'Quantity': [10, 5, 2, np.nan, 8],  # Another NaN value
        'Sector': ['Tech', 'Tech', 'Tech', 'Consumer Discretionary', 'Automotive']}
df = pd.DataFrame(data)
print(df)

# Checking for missing values (Boolean mask)
print(df.isnull())  # Returns True for NaN values, False otherwise
print(df.isna().sum()) # count how many NaN values per column - useful for quick overview

# Filling missing values (Imputation)
df_filled = df.fillna(0)  # Fill with 0 - be careful with this, it might not be appropriate for all columns
print(df_filled)

df_filled_mean = df.fillna(df['Price'].mean()) # Fill with column mean - often a better strategy for numerical data
print(df_filled_mean)

df_filled_median = df.fillna(df['Price'].median()) # Fill with column median - robust to outliers
print(df_filled_median)

df_filled_ffill = df.fillna(method='ffill')  # Forward fill (use the previous value) - useful for time series data
print(df_filled_ffill)

df_filled_bfill = df.fillna(method='bfill')  # Backward fill (use the next value) - useful for time series data
print(df_filled_bfill)

# Interpolation (for numerical data with a trend)
df_interpolated = df.interpolate()  # Linear interpolation by default
print(df_interpolated)

# Dropping rows with missing values (use with caution)
df_dropped = df.dropna()  # Drop any row with at least one NaN value - can lead to significant data loss
print(df_dropped)

df_dropped_subset = df.dropna(subset=['Price']) # Drop rows where Price is NaN - more targeted approach
print(df_dropped_subset) 
```

## Removing Duplicates


```python
# Adding a duplicate row
df = pd.concat([df, pd.DataFrame([{'Ticker': 'AAPL', 'Price': 150.25, 'Quantity': 10, 'Sector': 'Tech'}])], ignore_index=True)
print(df)

# Removing duplicates
df_no_duplicates = df.drop_duplicates()  # Removes all completely identical rows
print(df_no_duplicates)

# Removing duplicates based on specific columns
df_no_duplicates_ticker = df.drop_duplicates(subset=['Ticker']) # Keep only the first occurrence of each ticker
print(df_no_duplicates_ticker)

# Keep the last occurence of the duplicated rows
df_no_duplicates_last = df.drop_duplicates(subset=['Ticker'], keep='last') # Keep the last occurrence of each ticker
print(df_no_duplicates_last)

# Removing duplicates and keeping specific columns
df_no_duplicates_subset = df.drop_duplicates(subset=['Ticker', 'Sector']) # Consider both ticker and sector when identifying duplicates
print(df_no_duplicates_subset)
```

## Cleaning Financial Data: A Practical Example 


```python
import pandas as pd
import numpy as np  # Needed for NaN values

# Sample financial data (with some issues)
data = {
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05', '2024-01-06'],
    'Ticker': ['AAPL', 'MSFT', 'AAPL', 'GOOG', 'MSFT', 'AAPL'],
    'Price': ['150.25', '250.75', '155', '2000', '252.50', '155.50'],  # Price as string
    'Volume': ['10000', '5000', '7500', '2000', '6000', '8000'],  # Volume as string
    'Revenue': ['10M', '5M', '7.5M', '2M', '6M', '9M'],  # Revenue with units
    'Sector': ['Tech', 'Tech', 'Tech', 'Tech', 'Tech', 'Tech'],
    'Currency': ['USD', 'USD', 'USD', 'USD', 'USD', 'USD']  # Currency information
}

df = pd.DataFrame(data)
print("\nOriginal DataFrame:\n", df)

# Convert Price and Volume to numeric
df['Price'] = pd.to_numeric(df['Price'])
df['Volume'] = pd.to_numeric(df['Volume'])

# Clean Revenue (remove units and convert to numeric)
def clean_revenue(revenue):
    try:
        if isinstance(revenue, str):  # Ensure it's a string
            if 'M' in revenue:
                return float(revenue.replace('M', '')) * 1_000_000
            elif 'K' in revenue:
                return float(revenue.replace('K', '')) * 1_000
            else:
                return float(revenue)  # Handle pure numeric values
        return float(revenue)  # If it's already a number
    except ValueError:
        return np.nan  # Return NaN for unprocessable values

df['Revenue'] = df['Revenue'].apply(clean_revenue)

print("\nFinancial data after cleaning:\n", df)
```

## Merging: Combining DataFrames
### Examples of Merging


```python
import pandas as pd

# Sample DataFrames with values
df1 = pd.DataFrame({'Ticker': ['AAPL', 'MSFT', 'GOOG'], 'Price': [150, 250, 2800]})
df2 = pd.DataFrame({'Ticker': ['AAPL', 'MSFT', 'AMZN'], 'Quantity': [100, 200, 300]})

# Inner Join
inner_merged = pd.merge(df1, df2, on='Ticker', how='inner')
print("Inner Join:\n", inner_merged)

# Left Join
left_merged = pd.merge(df1, df2, on='Ticker', how='left')
print("\nLeft Join:\n", left_merged)

# Right Join
right_merged = pd.merge(df1, df2, on='Ticker', how='right')
print("\nRight Join:\n", right_merged)

# Outer Join
outer_merged = pd.merge(df1, df2, on='Ticker', how='outer')
print("\nOuter Join:\n", outer_merged)
```

```python
# Calculate daily returns
df['Daily Return'] = df['Close'].pct_change()
print("\nDaily Returns:\n", df['Daily Return'])

# Calculate cumulative returns
df['Cumulative Return'] = (1 + df['Daily Return']).cumprod() - 1
print("\nCumulative Returns:\n", df['Cumulative Return'])

# Calculate rolling statistics (e.g., 3-day rolling mean of closing price)
df['Rolling Mean'] = df['Close'].rolling(window=3).mean()
print("\n3-Day Rolling Mean:\n", df['Rolling Mean']) 
```

## Automating Excel Tasks: Unleashing the Power of Python

```python
import openpyxl
from openpyxl.chart import BarChart, Reference
from openpyxl.formatting.rule import ColorScaleRule
import csv

# --- 1. Create a sample CSV file ---
with open("data.csv", "w", newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(["Category", "Amount", "Sales"])
    writer.writerow(["A", 100, 150])
    writer.writerow(["B", 200, 250])
    writer.writerow(["C", 300, 350])
print("CSV file created successfully.")

# --- 2. Create Workbook and Read CSV into Sheet ---
workbook = openpyxl.Workbook()
sheet = workbook.active
sheet.title = "CSV Data"

with open("data.csv", "r") as csv_file:
    reader = csv.reader(csv_file)
    headers = next(reader)  # Read headers
    sheet.append(headers)   # Add headers to sheet
    for row in reader:
        # Convert numeric columns (assumed columns 2 & 3) to float
        new_row = []
        for idx, value in enumerate(row):
            if idx == 0:  # Category remains a string
                new_row.append(value)
            else:
                try:
                    new_row.append(float(value))
                except ValueError:
                    new_row.append(value)
        sheet.append(new_row)
print("CSV data added to sheet.")

# --- 3. Add Formula (Sum of Amount and Sales for Last Row) ---
last_row = sheet.max_row
sheet[f'D{last_row}'] = f'=SUM(B{last_row}:C{last_row})'
sheet['D1'] = "Total"  # Header for Total column

# --- 4. Create a Bar Chart ---
chart = BarChart()
chart.type = "col"  # Column chart
chart.style = 10
chart.title = "Sales Data"
chart.y_axis.title = "Sales Amount"
chart.x_axis.title = "Category"

# --- Set Correct Data Ranges ---
# Data: Include header row for titles_from_data=True (columns 2 and 3)
data = Reference(sheet, min_col=2, min_row=1, max_col=3, max_row=last_row)
# Categories: Only the category names (exclude header)
categories = Reference(sheet, min_col=1, min_row=2, max_row=last_row)

chart.add_data(data, titles_from_data=True)
chart.set_categories(categories)
chart.legend.position = "r"

# Place the chart in the sheet starting at cell E2
sheet.add_chart(chart, "E2")

# --- 5. Add Conditional Formatting (Color Scale) for Numeric Columns ---
rule = ColorScaleRule(
    start_type='min', start_color='FFAAAA',  # Light Red
    mid_type='percentile', mid_value=50, mid_color='FFFFAA',  # Yellow
    end_type='max', end_color='AAFFAA'  # Light Green
)
sheet.conditional_formatting.add(f'B2:C{last_row}', rule)

# --- 6. Save Workbook ---
workbook.save("automated_excel_file.xlsx")
print("Workbook saved successfully as 'automated_excel_file.xlsx'.") 
```

# Chapter 14.	Building a Simple Financial Application: Portfolio Tracker
## Building the Application


```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# Portfolio data. This is a sample data.
portfolio_data = {
    'Ticker': ['AAPL', 'MSFT', 'GOOG', 'AMZN', 'TSLA', 'NVDA'],
    'Quantity': [10, 5, 2, 3, 8, 4],
    'Purchase Price': [150.00, 250.00, 2000.00, 100.00, 700.00, 300.00],
    'Purchase Date': ['2023-01-15', '2023-02-20', '2023-03-10', '2023-04-05', '2023-05-18', '2023-06-22']
}

portfolio = pd.DataFrame(portfolio_data)

# Write the data to portfolio.csv
portfolio.to_csv("portfolio.csv", index=False)

# Load the portfolio data
portfolio = pd.read_csv("portfolio.csv")
portfolio['Purchase Date'] = pd.to_datetime(portfolio['Purchase Date'])

# Fetch current stock prices using yfinance and yf.Ticker
def get_current_price(ticker):
    try:
        ticker_obj = yf.Ticker(ticker)
        hist = ticker_obj.history(period="1d")
        if not hist.empty:
            return float(hist['Close'].iloc[-1])
        else:
            print(f"No data for {ticker}")
            return None
    except Exception as e:
        print(f"Error fetching price for {ticker}: {e}")
        return None

portfolio['Current Price'] = portfolio['Ticker'].apply(get_current_price)

# Calculate portfolio value and gain/loss
portfolio['Value'] = portfolio['Quantity'] * portfolio['Current Price']
portfolio['Gain/Loss'] = portfolio['Value'] - (portfolio['Quantity'] * portfolio['Purchase Price'])

# Calculate total portfolio value
total_value = portfolio['Value'].sum()
print("Total Portfolio Value:", total_value)

# Asset allocation (percentage of each stock in the portfolio)
asset_allocation = portfolio.groupby('Ticker')['Value'].sum() / total_value
print("\nAsset Allocation:\n", asset_allocation)

# Plot portfolio value over time
portfolio = portfolio.sort_values(by='Purchase Date')
plt.plot(portfolio['Purchase Date'], portfolio['Value'], marker='o')
plt.xlabel("Purchase Date")
plt.ylabel("Value")
plt.title("Portfolio Value Over Time")
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()

# Display the portfolio DataFrame with all calculated columns
print("\nPortfolio Details:\n", portfolio)
```
